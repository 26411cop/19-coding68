```
import time
from IPython.display import clear_output

total = int(input("‚è±Ô∏è ‡πÉ‡∏™‡πà‡πÄ‡∏ß‡∏•‡∏≤‡πÇ‡∏ü‡∏Å‡∏±‡∏™ (‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ): "))

for remaining in range(total, -1, -1):
    clear_output(wait=True)

    percent = (total - remaining) / total
    bar_len = 20
    filled = int(bar_len * percent)
    bar = "‚ñà" * filled + "‚ñë" * (bar_len - filled)

    m, s = divmod(remaining, 60)

    print("üî• FOCUS TIME üî•")
    print(f"[{bar}] {int(percent*100)}%")
    print(f"‚è≥ ‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏ß‡∏•‡∏≤ {m:02d}:{s:02d}")

    time.sleep(1)

print("\nüéâ ‡∏´‡∏°‡∏î‡πÄ‡∏ß‡∏•‡∏≤‡πÅ‡∏•‡πâ‡∏ß! ‡∏û‡∏±‡∏Å‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢ üëè")
```

```
import tkinter as tk
from tkinter import ttk
import time

# ---------- Logic ----------
running = False
paused = False
total_time = 0
remaining = 0

def start():
    global running, paused, total_time, remaining
    if not running:
        total_time = int(entry.get())
        remaining = total_time
        progress["maximum"] = total_time
        running = True
        paused = False
        update()

def pause():
    global paused
    paused = not paused
    pause_btn.config(text="‡∏ó‡∏≥‡∏ï‡πà‡∏≠" if paused else "‡∏´‡∏¢‡∏∏‡∏î‡∏û‡∏±‡∏Å")

def reset():
    global running, paused
    running = False
    paused = False
    label.config(text="0 ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ")
    progress["value"] = 0
    pause_btn.config(text="‡∏´‡∏¢‡∏∏‡∏î‡∏û‡∏±‡∏Å")

def update():
    global remaining, running
    if running and not paused:
        if remaining >= 0:
            label.config(text=f"{remaining} ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ")
            progress["value"] = total_time - remaining
            remaining -= 1
            root.after(1000, update)
        else:
            label.config(text="‚è∞ ‡∏´‡∏°‡∏î‡πÄ‡∏ß‡∏•‡∏≤‡πÅ‡∏•‡πâ‡∏ß!")
            running = False

# ---------- UI ----------
root = tk.Tk()
root.title("üî• Focus Timer")
root.geometry("500x350")
root.configure(bg="#1e1e2e")

style = ttk.Style()
style.theme_use("default")
style.configure(
    "TProgressbar",
    thickness=25,
    troughcolor="#3b3b4f",
    background="#00ffcc"
)

title = tk.Label(
    root,
    text="FOCUS TIMER",
    font=("Arial", 22, "bold"),
    fg="#00ffcc",
    bg="#1e1e2e"
)
title.pack(pady=10)

tk.Label(
    root,
    text="‡πÉ‡∏™‡πà‡πÄ‡∏ß‡∏•‡∏≤ (‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ)",
    fg="white",
    bg="#1e1e2e"
).pack()

entry = tk.Entry(root, font=("Arial", 14), justify="center")
entry.pack(pady=5)
entry.insert(0, "60")

label = tk.Label(
    root,
    text="0 ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ",
    font=("Arial", 32, "bold"),
    fg="white",
    bg="#1e1e2e"
)
label.pack(pady=20)

progress = ttk.Progressbar(root, length=400)
progress.pack(pady=10)

btn_frame = tk.Frame(root, bg="#1e1e2e")
btn_frame.pack(pady=20)

start_btn = tk.Button(
    btn_frame, text="‡πÄ‡∏£‡∏¥‡πà‡∏°",
    width=10, bg="#00ffcc", fg="black",
    command=start
)
pause_btn = tk.Button(
    btn_frame, text="‡∏´‡∏¢‡∏∏‡∏î‡∏û‡∏±‡∏Å",
    width=10, bg="#ffaa00", fg="black",
    command=pause
)
reset_btn = tk.Button(
    btn_frame, text="‡∏£‡∏µ‡∏™‡∏ï‡∏≤‡∏£‡πå‡∏ï",
    width=10, bg="#ff5555", fg="white",
    command=reset
)

start_btn.grid(row=0, column=0, padx=5)
pause_btn.grid(row=0, column=1, padx=5)
reset_btn.grid(row=0, column=2, padx=5)

root.mainloop()
```
